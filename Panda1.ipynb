```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv(
    "QS_World_University_Rankings_2025_Top_global_universities (1).csv",
    encoding="latin1"
)
```

```python
print("\nFIRST 5 ROWS")
print(df.head())

print("\nLAST 5 ROWS")
print(df.tail())

print("\nDATASET SHAPE")
print(df.shape)

print("\nCOLUMN NAMES")
print(df.columns)

print("\nDATA TYPES")
print(df.dtypes)

print("\nFULL INFORMATION")
print(df.info())
```

```python
print("\nMISSING VALUES IN EACH COLUMN")
print(df.isnull().sum())

print("\nTOTAL MISSING VALUES")
print(df.isnull().sum().sum())
```

```python
print("\nTOTAL DUPLICATE ROWS")
print(df.duplicated().sum())

```

```python
df = df.drop_duplicates()

print("\nDUPLICATES REMOVED")
```

```python
# column names cleaning

df.columns = df.columns.str.strip()

# text data cleaning

text_cols = df.select_dtypes(include=["object", "string"]).columns

for col in text_cols:
    df[col] = df[col].astype(str).str.strip()

print("\nEXTRA SPACES REMOVED")

```

```python
# numeric columns

numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())



# text columns

text_cols = df.select_dtypes(include=["object", "string"]).columns

for col in text_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("\nMISSING VALUES FILLED")
```

```python
print("\nMISSING VALUES AFTER CLEANING")
print(df.isnull().sum())

print("\nTOTAL MISSING VALUES AFTER CLEANING")
print(df.isnull().sum().sum())

```

```python
for col in df.columns:

    try:
        df[col] = pd.to_numeric(df[col])

    except:
        pass

print("\nPOSSIBLE NUMERIC COLUMNS CONVERTED")
```

```python
print("\nFINAL DATASET INFORMATION")
print(df.info())

print("\nFINAL DATA TYPES")
print(df.dtypes)

print("\nFINAL SHAPE")
print(df.shape)
```

```python
print("\nSTATISTICAL SUMMARY")
print(df.describe())
```

```python
df.to_csv(
    "cleaned_dataset.csv",
    index=False
)

print("\nCLEANED DATASET SAVED SUCCESSFULLY")
```
