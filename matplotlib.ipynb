```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x, y)

plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales Growth")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [15,30,45,60]

plt.plot(x, y,
         color='green',
         linestyle='--',
         linewidth=2,
         marker='o',
         markersize=10)

plt.xlabel("Month")
plt.ylabel("Profit")
plt.title("Company Profit")

plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]

y1 = [10,20,30,40]
y2 = [40,30,20,10]

plt.plot(x, y1, label="Profit")
plt.plot(x, y2, label="Loss")

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

subjects = ["Math","Science","English"]
marks = [90,85,95]

plt.bar(subjects, marks)

plt.show()
```

```python
import matplotlib.pyplot as plt

marks = [50,55,60,60,65,70,75,80,85,90]

plt.hist(marks)

plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [40,30,20,10]
labels = ["Python","Java","C++","JS"]

plt.pie(sizes,
        labels=labels)

plt.show()
```

```python
plt.pie(sizes,
        labels=labels,
        autopct='%1.1f%%')
```

```python
explode = [0.1,0,0,0]

plt.pie(sizes,
        labels=labels,
        explode=explode)
```
